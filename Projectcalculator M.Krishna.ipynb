```python
from tkinter import *

window = Tk()
window.title('Simple Calculator')
window.iconbitmap('J:/PYTHON/calc.ico')

e = Entry(width=45, borderwidth=5)
e.grid(row=0, column=0, columnspan=4)

def Buttonclick(number):
    current=e.get()
    e.delete(0,END)
    e.insert(1,str(current)+str(number))

def Buttonclear():
    e.delete(0,END)

def ButtonAdd():
    global fnum,math
    math=1
    first_number=e.get()
    e.delete(0,END)
    fnum=int(first_number)

def ButtonSub():
    global fnum,math
    math=2
    first_number=e.get()
    e.delete(0,END)
    fnum=int(first_number)

def ButtonMul():
    global fnum,math
    math=3
    first_number=e.get()
    e.delete(0,END)
    fnum=int(first_number)

def ButtonDiv():
    global fnum,math
    math=4
    first_number=e.get()
    e.delete(0,END)
    fnum=int(first_number)

def ButtonEqual():
    second_number=e.get()
    e.delete(0,END)
    snum=int(second_number)

    if math == 1:
        e.insert(1,fnum+snum)

    elif math == 2:
        e.insert(1,fnum-snum)

    elif math == 3:
        e.insert(1,fnum*snum)

    else:
        e.insert(1,fnum/snum)

button0 = Button(text=0, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(0)).grid(row=4, column=0)
button1 = Button(text=1, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(1)).grid(row=3, column=0)
button2 = Button(text=2, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(2)).grid(row=3, column=1)
button3 = Button(text=3, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(3)).grid(row=3, column=2)
button4 = Button(text=4, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(4)).grid(row=2, column=0)
button5 = Button(text=5, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(5)).grid(row=2, column=1)
button6 = Button(text=6, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(6)).grid(row=2, column=2)
button7 = Button(text=7, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(7)).grid(row=1, column=0)
button8 = Button(text=8, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(8)).grid(row=1, column=1)
button9 = Button(text=9, fg="red", bg="black", padx=40, pady=20,font="bold",command=lambda:Buttonclick(9)).grid(row=1, column=2)


buttonadd = Button(text="+", fg="blue", bg="black", padx=38, pady=20,font="bold",command=ButtonAdd).grid(row=1, column=3)
buttonsub = Button(text="-", fg="blue", bg="black", padx=40, pady=20,font="bold",command=ButtonSub).grid(row=2, column=3)
buttonmul = Button(text="*", fg="blue", bg="black", padx=40, pady=20,font="bold",command=ButtonMul).grid(row=3, column=3)
buttondiv = Button(text="/", fg="blue", bg="black", padx=40, pady=20,font="bold",command=ButtonDiv).grid(row=4, column=3)

buttonequal = Button(text="=", fg="blue", bg="black", padx=91, pady=20,font="bold",command=ButtonEqual ).grid(row=4, column=1, columnspan=2)
buttonclear = Button(text='Clear', fg="green", bg="black",padx=176, pady=20,font="bold",command=Buttonclear ).grid(row=5, column=0, columnspan=4)

window.mainloop()
```
